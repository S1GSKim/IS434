snscrape

Ref: https://github.com/JustAnotherArchivist/snscrape

```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

Scraping a specific Twitter user‚Äôs tweets

Let's use this handle: @govsingapore

Using the code, we can scrape 100 tweets from Twitter user @govsingapore
We then pulled the DateTime, tweet id, text, and username attributes from the tweet objects.

```python
# Creating list to append tweet data 
tweets_list1 = []

# Using TwitterSearchScraper to scrape data and append tweets to list
handle = 'govsingapore'

# declare a username 
for i, tweet in enumerate(sntwitter.TwitterSearchScraper('from:' + handle).get_items()): 
    if i > 100: #number of tweets you want to scrape
        break
        
    #declare the attributes to be returned
    tweets_list1.append([tweet.date, tweet.id, tweet.content, tweet.user.username])
    
# Creating a dataframe from the tweets list above 
tweets_df1 = pd.DataFrame(tweets_list1, columns=['Datetime', 'Tweet Id', 'Text', 'Username'])
```

```python
tweets_df1
```

Scraping tweets from a text search query

Using the code below, we are scraping 5000 tweets between:  2022-01-01 until:2023-01-31,
with the keywords 'covid-19'.

I then pulled attributes DateTime, tweet id, text, and username from the tweet objects.

```python
# Creating list to append tweet data to
tweets_list2 = []

# search terms
search_terms = 'covid-19'

# Using TwitterSearchScraper to scrape data and append tweets to list
for i, tweet in enumerate(sntwitter.TwitterSearchScraper(search_terms + ' since:2022-01-01 until:2023-01-31').get_items()):
    if i > 5000:
        break
    #declare the attributes to be returned
    tweets_list2.append([tweet.date, tweet.id, tweet.content, tweet.user.username])

# Creating a dataframe from the tweets list above
tweets_df2 = pd.DataFrame(tweets_list2, columns=['Datetime', 'Tweet Id', 'Text', 'Username'])
```

```python
tweets_df2
```

```python
tweets_df2.to_csv('covid_tweets.csv', index=False, encoding='utf-8')
```
